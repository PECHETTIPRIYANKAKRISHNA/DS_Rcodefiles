```python
from scipy import stats
import pandas as pd
import numpy as np
bemldata1=pd.read_csv("C:\\Users\\PRIYANKA\\Downloads\\BEML.csv")
bemldata1[0:3]
```

```python
bemldata2=bemldata1.set_index(pd.DatetimeIndex(bemldata1['Date']))
bemldata3=bemldata2[['Date','Close']]
bemldata3
```

```python
bemldata3['gain']=bemldata3.Close.pct_change(periods=1)
bemldata3
```

```python
bemldata_ci=stats.norm.interval(0.95,loc=bemldata3.gain.mean(),scale=bemldata3.gain.std())
bemldata_ci
print("gain at 95% confidence interval is:",np.round(bemldata_ci,4))
```
