```python
f,ax=plt.subplots(figsize=(8,4))
sns.boxplot(x='Survived',y='Age',data=traindata)
plt.show()
```

```python
sns.barplot(x='Pclass',y='Survived',data=traindata)
```

```python
sns.barplot(x='Pclass',y='Survived',hue="Sex",data=traindata)
```

```python
sns.barplot(x='SibSp',y='Survived',data=traindata)
```

```python
sns.barplot(x='Parch',y='Survived',data=traindata)
```

```python
survived=traindata.loc[traindata['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
notsurvived=traindata.loc[traindata['Survived']==0,"Age"].dropna()
sns.distplot(notsurvived)
```

```python
sns.pairplot(traindata.dropna())
```

```python
sns.pairplot(traindata.dropna(),hue='Sex')
```

```python
#Pclass and Survive
grid=sns.FacetGrid(traindata,col='Survived',row='Pclass',size=2.5,aspect=1.5)
grid.map(plt.hist,'Age',alpha=0.5,bins=10)
grid.add_legend()
```

```python
figbi,axesbi=plt.subplots(2,4,figsize=(16,10))
traindata.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
traindata.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
traindata.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
traindata.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
traindata.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x='Survived',y='Age',data=traindata,ax=axesbi[1,1])
sns.boxplot(x='Survived',y='Fare',data=traindata,ax=axesbi[1,2])

```
